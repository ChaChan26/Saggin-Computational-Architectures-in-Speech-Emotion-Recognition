# Data Validation & Cleaning

This notebook performs:
1. Verification of the tabular features CSV file (`all_emotions.csv`).
2. Label class verification and distribution.
3. Clean-up of completely empty rows or duplicate records to prepare the final clean dataset.


```python
# ============================================================
# Library Imports
# ============================================================
import os
import pandas as pd
```

## File Path Resolution

We configure path variables to locate the audio features dataset.


```python
# ============================================================
# Path Configuration and Global Constants
# ============================================================
_base = os.getcwd()
_project_root = os.path.dirname(_base) if os.path.basename(_base).lower() == "training" else _base
ALL_EMOTIONS_CSV = os.path.normpath(os.path.join(_project_root, "dataset", "all_emotions.csv"))
if not os.path.isfile(ALL_EMOTIONS_CSV):
    fallback_csv = os.path.normpath(os.path.join(_project_root, "all_emotions.csv"))
    if os.path.isfile(fallback_csv):
        ALL_EMOTIONS_CSV = fallback_csv

print("Dataset path:", ALL_EMOTIONS_CSV)
print("Dataset file exists:", os.path.isfile(ALL_EMOTIONS_CSV))
```

## Ingest and Explore Labels

We load the raw CSV and print the distribution of classes before cleaning.


```python
# ============================================================
# Data Loading and Inspection
# ============================================================
df = pd.read_csv(ALL_EMOTIONS_CSV)
target_col = "label" if "label" in df.columns else "Label"

print("Raw Dataset Size:", len(df))
print("\n--- Distribution of Classes ---")
print(df[target_col].value_counts(dropna=False))
```

## Remove Null Targets and Duplicate Records

We drop records where the target class is null or contains 'nan' values, remove duplicate feature entries, and save the clean dataset back to the CSV.


```python
# ============================================================
# Data Cleaning: Removing Nulls and Duplicates
# ============================================================
df_cleaned = df.dropna(subset=[target_col]).copy()
df_cleaned = df_cleaned[df_cleaned[target_col].astype(str).str.strip().str.lower() != "nan"]

# Remove duplicate feature rows
df_cleaned = df_cleaned.drop_duplicates()

# Save cleaned dataset back to the CSV file
df_cleaned.to_csv(ALL_EMOTIONS_CSV, index=False)

print("Cleaned Dataset Size:", len(df_cleaned))
print("Removed duplicates/nulls count:", len(df) - len(df_cleaned))
print("\n--- Cleaned Label Counts ---")
print(df_cleaned[target_col].value_counts(dropna=False))
```
