```python
import os
import pandas as pd

_base = os.getcwd()
_project_root = (
    os.path.dirname(_base) if os.path.basename(_base).lower() == "training" else _base
)
ALL_EMOTIONS_CSV = os.path.normpath(os.path.join(_project_root, "dataset", "all_emotions.csv"))

df = pd.read_csv(ALL_EMOTIONS_CSV)
df = df.dropna(how="all")
df = df.drop_duplicates()
df.to_csv(ALL_EMOTIONS_CSV, index=False)

print(df["label"].value_counts())
```
